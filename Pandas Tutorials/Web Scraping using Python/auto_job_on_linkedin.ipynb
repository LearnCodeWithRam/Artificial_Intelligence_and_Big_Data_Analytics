
# LinkedIn Auto Apply Script

## Step 1: Set Up Your Python Script
Create a Python script, for example, `linkedin_auto_apply.py`. Start by importing the necessary libraries and setting up the WebDriver:


```python

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

## Step 2: Log In to LinkedIn
Use Selenium to navigate to LinkedInâ€™s login page and log in with your credentials:

```python

# Configure WebDriver
driver = webdriver.Chrome()

# Log in to LinkedIn
driver.get('https://www.linkedin.com/login')
email = driver.find_element(By.ID, 'username')
password = driver.find_element(By.ID, 'password')
```

```python
email.send_keys('demo@gmail.com')
password.send_keys('**********')
password.send_keys(Keys.RETURN)

# Wait until logged in
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, 'global-nav-typeahead'))
)
```

## Step 3: Search for Jobs
Navigate to the LinkedIn jobs page and perform a job search:

```python

# Go to the jobs page
driver.get('https://www.linkedin.com/jobs/')

# Search for jobs
search_box = driver.find_element(By.CLASS_NAME, 'jobs-search-box__text-input')
search_box.send_keys('Software Engineer')  # You can replace 'Software Engineer' with your desired job title
search_box.send_keys(Keys.RETURN)

# Wait until search results are loaded
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, 'job-card-container'))
)
```

```python

# Loop through job listings and apply
jobs = driver.find_elements(By.CLASS_NAME, 'job-card-container')
for job in jobs:
    try:
        # Click the job to open the details
        job.click()
        time.sleep(2)  # Wait for the page to load

        # Click the Easy Apply button
        easy_apply_button = driver.find_element(By.XPATH, "//button[contains(@class, 'jobs-apply-button')]")
        easy_apply_button.click()
        time.sleep(2)

        # Handle the application form (this may need to be customized for each job)
        submit_button = driver.find_element(By.XPATH, "//button[@type='submit' and @aria-label='Submit application']")
        submit_button.click()
        time.sleep(2)  # Wait for the application to be submitted

    except Exception as e:
        print(f"Could not apply for job: {e}")
        continue
```

## Step 4: Handle Pagination (Optional)
If you want to apply to more jobs beyond the first page of results, handle pagination:

```python

while True:
    try:
        next_button = driver.find_element(By.XPATH, "//button[@aria-label='Next']")
        next_button.click()
        time.sleep(5)  # Wait for the next page to load

        # Repeat the job application loop for the next page of jobs
        jobs = driver.find_elements(By.CLASS_NAME, 'job-card-container')
        for job in jobs:
            try:
                job.click()
                time.sleep(2)
                easy_apply_button = driver.find_element(By.XPATH, "//button[contains(@class, 'jobs-apply-button')]")
                easy_apply_button.click()
                time.sleep(2)
                submit_button = driver.find_element(By.XPATH, "//button[@type='submit' and @aria-label='Submit application']")
                submit_button.click()
                time.sleep(2)
            except Exception as e:
                print(f"Could not apply for job: {e}")
                continue

    except:
        print("No more pages.")
        break
```

```python
driver.quit()
```
