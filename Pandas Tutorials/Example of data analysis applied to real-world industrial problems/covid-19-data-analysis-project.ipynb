```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
covid_df.info()
```

```python
#Active cases
covid_df['Active_cases']=covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

```python
covid_df['State/UnionTerritory']=covid_df['State/UnionTerritory'].replace('Maharashtra***',"Maharashtra")
covid_df['State/UnionTerritory']=covid_df['State/UnionTerritory'].replace('Bihar****',"Bihar")
covid_df['State/UnionTerritory']=covid_df['State/UnionTerritory'].replace('Madhya Pradesh***',"Madhya Pradesh")
covid_df['State/UnionTerritory']=covid_df['State/UnionTerritory'].replace('Karanataka',"Karnataka")
```

```python
#pivot table
statewise_data=pd.pivot_table(covid_df, values=['Confirmed','Deaths','Cured'],index="State/UnionTerritory", aggfunc=max)
```

```python
statewise_data['Recovery Rate']=statewise_data['Cured']*100/statewise_data['Confirmed']
```

```python
statewise_data['Mortality Rate']=statewise_data['Deaths']*100/statewise_data['Confirmed']
```

```python
statewise_data=statewise_data.sort_values(by="Confirmed",ascending=False)
```

```python
statewise_data.style.background_gradient(cmap="CMRmap")
```

```python
# top 10 active cases states
top10ActiveCases=covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
print(top10ActiveCases)
```

```python
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 states with most Active Cases  in India",size=22)
ax=sns.barplot(data=top10ActiveCases.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel("State/UnionTerritory")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# top states with highest number of deaths
top10Deaths=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,10))
plt.title("Top 10 states with most Deaths in India",size=22)
ax=sns.barplot(data=top10Deaths.iloc[:10],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel("States")
plt.ylabel("Total Deaths")
plt.show()
```

```python
#Growth Trend
fig=plt.figure(figsize=(12,6))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date',y='Active_cases',hue='State/UnionTerritory')
#ax.set_title("Top 5 Affected States in India",size=16)
```

```python
#Male vs Female Vaccination

import matplotlib.pyplot as plt

# Summing up the vaccination counts
male = vaccination['Male(Individuals Vaccinated)'].sum()
female = vaccination['Female(Individuals Vaccinated)'].sum()

# Data for the pie chart
sizes = [male, female]
labels = ["Male", "Female"]

# Creating the pie chart
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Male vs Female Vaccination")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(title="Two Genders:", bbox_to_anchor=(1.05, 1), loc='upper left')
# Showing the pie chart
plt.show()

```

# Comparison of Seaborn and Matplotlib

Both Seaborn and Matplotlib are powerful Python libraries for data visualization, but they have some key differences:

## Matplotlib
- **Flexibility:** Highly customizable and versatile, allowing for detailed control over every aspect of a plot.
- **Complexity:** Requires more code to create complex visualizations.
- **Use Cases:** Ideal for creating a wide range of static, animated, and interactive plots.
- **Integration:** Works well with NumPy and Pandas for data manipulation.

## Seaborn
- **Built on Matplotlib:** Simplifies the process of creating attractive and informative statistical graphics.
- **Ease of Use:** Provides a high-level interface for drawing attractive and informative statistical graphics with less code.
- **Default Styles:** Comes with beautiful default themes and color palettes.
- **Specialized Plots:** Offers specialized functions for statistical plots like violin plots, box plots, and heatmaps.

## Comparison
- **Syntax:** Seaborn has a simpler and more intuitive syntax compared to Matplotlib.
- **Customization:** Matplotlib offers more customization options, while Seaborn focuses on ease of use and aesthetics.
- **Plot Types:** Both libraries can create similar types of plots, but Seaborn is particularly strong in statistical visualizations.


To create a figure with two rows and three columns, incorporating the various plots you've provided, you can use the matplotlib.pyplot.subplots function to set up the grid and then plot each of your figures in the appropriate subplot. Here's how you can organize and plot all the figures into a single figure with a 2x3 grid:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 2x3 grid for subplots
fig, axes = plt.subplots(2, 3, figsize=(30, 20))

# Plot 0: Top 10 states with most Active Cases in India
sns.barplot(data=top10ActiveCases.iloc[:10], y='Active_cases', x='State/UnionTerritory', linewidth=2, edgecolor='black', ax=axes[0, 0])
axes[0, 0].set_title("Top 10 states with most Active Cases in India", size=22)
axes[0, 0].set_xlabel("State/UnionTerritory")
axes[0, 0].set_ylabel("Total Active Cases")

# Plot 1: Top 10 states with most Deaths in India
sns.barplot(data=top10Deaths.iloc[:10], y='Deaths', x='State/UnionTerritory', linewidth=2, edgecolor='black', ax=axes[0, 1])
axes[0, 1].set_title("Top 10 states with most Deaths in India", size=22)
axes[0, 1].set_xlabel("States")
axes[0, 1].set_ylabel("Total Deaths")

# Plot 2: Growth Trend
sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])], x='Date', y='Active_cases', hue='State/UnionTerritory', ax=axes[0, 2])
axes[0, 2].set_title("Growth Trend in Selected States", size=22)
axes[0, 2].set_xlabel("Date")
axes[0, 2].set_ylabel("Active Cases")

# Plot 3: Pie chart for Male vs Female Vaccination
sizes = [male, female]
labels = ["Male", "Female"]
axes[1, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axes[1, 0].set_title("Male vs Female Vaccination", size=16)
axes[1, 0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Plot 4: Top 5 Vaccinated States in India
sns.barplot(data=max_vacc.iloc[:5], y='Total', x=max_vacc.index[:5], linewidth=2, edgecolor='black', ax=axes[1, 1])
axes[1, 1].set_title("Top 5 Vaccinated States in India", size=16)
axes[1, 1].set_xlabel("States")
axes[1, 1].set_ylabel("Total Vaccinations")

# Plot 5: Least 5 Vaccinated States in India
sns.barplot(data=min_vacc.iloc[:5], y='Total', x=min_vacc.index[:5], linewidth=2, edgecolor='black', ax=axes[1, 2])
axes[1, 2].set_title("Least 5 Vaccinated States in India", size=16)
axes[1, 2].set_xlabel("States")
axes[1, 2].set_ylabel("Total Vaccinations")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

## Explanation:
- `plt.subplots(2, 3, figsize=(18, 12))`: Creates a figure with a grid of 2 rows and 3 columns.
- `ax=axes[i, j]`: Specifies the subplot location for each plot.
- `plt.tight_layout()`: Adjusts subplot parameters to give specified padding and prevent overlap.
